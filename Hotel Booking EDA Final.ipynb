```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)
```

```python
plt.title('Reservation Status Count', fontsize = 16)
plt.xlabel('Reservation Status', fontsize = 12)
plt.ylabel('Percantage', fontsize = 12)
plt.bar(["Not Cancelled","Cancelled"], df["is_canceled"].value_counts(), edgecolor='k', width = 0.7)
plt.tight_layout()
plt.show()
```

```python
resort_hotel = df[df["hotel"] == "Resort Hotel"]
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df["hotel"] == "City Hotel"]
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
plt.figure(figsize=(8,4))
axis1 = sns.countplot(x ='hotel', hue = 'is_canceled', data = df)
plt.title("Reservation Status of Both Hotels", fontsize = 16)
plt.xlabel("Hotels")
plt.ylabel("Number of reservtions")
plt.legend(['Not Canceled', 'Canceled'])
plt.show()
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
resort_hotel
```

```python
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel
```

```python
plt.figure(figsize = (20,8))
plt.title("Average Daily Rate in City and Resort Hotel", fontsize = 20)
plt.plot(resort_hotel.index,resort_hotel['adr'], label = "Resort Hotel")
plt.plot(city_hotel.index,city_hotel['adr'],label = "City Hotel")
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
```

```python
plt.figure(figsize = (16,8))
ax = sns.countplot(x='month', hue ='is_canceled', data = df, palette = 'bright')
plt.title("No of Reservation Status Per Month", fontsize=20)
plt.xlabel("Month")
plt.ylabel("Number of Reservations")
plt.legend(["Not Canceled", "Canceled"])

plt.show()
```

```python
canceled_data_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_data_adr.reset_index(inplace =True)
canceled_data_adr.sort_values('reservation_status_date', inplace = True)
```

```python
canceled_data_adr
```

```python
not_canceled_data = df[df["is_canceled"] == 0 ]
not_canceled_data_adr = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_data_adr.reset_index(inplace =True)
not_canceled_data_adr.sort_values('reservation_status_date', inplace = True)
not_canceled_data_adr
```

```python
plt.figure(figsize= (20,6))
plt.title('Average daily Rate of Canceled and Not Canceled')
plt.plot(not_canceled_data_adr['reservation_status_date'],not_canceled_data_adr['adr'], label = "Not Canceled")
plt.plot(canceled_data_adr['reservation_status_date'],canceled_data_adr['adr'], label = "Canceled")
plt.legend()
plt.show()
```
